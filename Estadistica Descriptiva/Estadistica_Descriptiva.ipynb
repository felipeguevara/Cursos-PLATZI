# Transformacion Lineal

```python
import timeit #para medir el tiempo de ejecución de los modelos
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model #datasets para descargar un modelo y linear_model para hacer una regresión lineal

X, y = datasets.load_diabetes(return_X_y=True) #carga el dataset
raw = X[:, None, 2] #transformación en las dimensiones para que se ajuste al formato de entrada del preprocesamiento
```

```python
#reglas de escalamiento lineal, aplicamos max-min
max_raw = max(raw)
#raw = datos crudos
min_raw = min(raw)
scaled = (2*raw - max_raw - min_raw)/(max_raw - min_raw)

# es importante tener una noción de los datos originales antes y después de escalarlos:
fig, axs = plt.subplots(2, 1, sharex=True)
axs[0].hist(raw)
axs[1].hist(scaled)
```

```python
# modelos para entrenamiento
def train_raw():
    linear_model.LinearRegression().fit(raw, y)

def train_scaled():
    linear_model.LinearRegression().fit(scaled, y)
```

```python
raw_time = timeit.timeit(train_raw, number=100) #repite la ejecución del código 100 veces y sobre eso calcula el tiempo
scaled_time = timeit.timeit(train_scaled, number=100)
print(f'train raw: {raw_time}')
print(f'train scaled: {scaled_time}')
```

**Se puede ver como al normalizar los datos, el algoritmo se vuelve más eficiente.**

# Transformacion No Lineal

```python
import timeit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
```

```python
raw = X[:,None,2]
"""Lo que quiere decir el None en este caso es algo asi como aumentar una dimension, 
ej: array([2,3,4]) se convierte en array([[2],[3],[4]]).
Finalmente si uno usa despues del None, una coma, se esta seleccionando que columna, si fuera 2, seleccionaria el numero _4_
Por Default es ":"
"""
```

```python
#Reglas de escalamiento
max_raw = max(raw)
min_raw = min(raw)

scaled_or_minmax = (2*raw - max_raw - min_raw)/(max_raw - min_raw)

fig, axs = plt.subplots(2,1, sharex = True) #It means that both graphics share the same xlabel

axs[0].hist(raw)
axs[1].hist(scaled_or_minmax)
```

```python
def train_raw():
    linear_model.LinearRegression().fit(raw,y)

def train_scaled():
    linear_model.LinearRegression().fit(scaled,y)
```

```python
raw_time = timeit.timeit(train_raw, number = 100)
scaled_time = timeit.timeit(train_scaled, number = 100)
print('train_raw : {}'.format(raw_time))
print('scaled_time : {}'.format(scaled_time))
```

# Analisis de componentes principales

```python
import timeit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
iris = sns.load_dataset('iris')
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
covariance_matrix = np.cov(scaled, rowvar = False)
covariance_matrix
```

```python
sns.pairplot(iris)
```

```python
sns.jointplot(x= iris.petal_length, y= iris.petal_width)
sns.jointplot(x= scaled[:,2], y= scaled[:,3])
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
variance_explained = []
for i in eigen_values:
    variance_explained.append((i/sum(eigen_values))*100)
print(variance_explained)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled)
```

```python
pca.explained_variance_ratio_
```

```python
reduced_scaled = pca.transform(scaled)
```

```python
reduced_scaled
```

```python
iris['pca_1'] = reduced_scaled[:,0]
iris['pca_2'] = reduced_scaled[:,1]
```

```python
iris
```

```python
sns.jointplot(data=iris, x=iris.pca_1, y=iris.pca_2, hue= iris.species)
```
