#### 2) Muestreo sistematico:

```python
def systematic_sampling(econdata, step):
    indexes = np.arange(0, len(econdata), step=step) #create an index array to get the index of the step elements in the df
    systematic_sample = econdata.iloc[indexes, :]
    return systematic_sample

systematic_sample = systematic_sampling(econdata, 3)
systematic_sample.head()
```

#### 3) Muestreo Estratificado

```python
print('Give us the proportion of each group')

econdata['estratificacion'] = econdata['delegacion'] + ',' + econdata['tipo']
(econdata['estratificacion'].value_counts()/len(econdata)).sort_values(ascending=False)
```

```python
def data_estratificada(econdata, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state=None):
    """Recibe un df con una columna previamente estratificada y sus elementos para ajustar un nuevo 
    df_estrat a tener la proporcion que deseamos de esa estratificacion"""
    
    df_estrat = pd.DataFrame(columns= econdata.columns)
    
    pos= -1
    for i in range(len(valores_estrat)):
        pos+=1
        if pos == len(valores_estrat) - 1:
            ratio_len = len(econdata) - len(df_estrat) #Para la ultima iteracion, el ratio va a hacer el total menos todas las prop_estrat que ya habiamos cargado
        else:
            ratio_len = int(len(econdata) * prop_estrat[i]) #Aca solo tomamos tantos clientes como proporcion necesitemos
        
        df_filtrado =  econdata[econdata[nombres_columnas_estrat] == valores_estrat[i]] #Se filtra por valores estrat actual i
        df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state)

        df_estrat = pd.concat([df_estrat, df_temp])
    return df_estrat
```

```python
valores_estrat = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel', 'Cuauhtémoc,Mercado', 'Venustiano Carranza,Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1, 0.1]
df_estrat = data_estratificada(econdata, 'estratificacion', valores_estrat, prop_estrat, random_state=42)
df_estrat
```

* Básicamente es muy útil cuando quieres indicar qué proporción de datos de una misma categoría quieres obtener.
* Obtener el mismo muestreo (con las mismas condiciones) gracias al random_state.
