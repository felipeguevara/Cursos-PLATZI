```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, '--')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
#axes.plot(x, y, '--')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, '--')
axes[1].plot(y, x, 'r--')
```

```python
fig, (axes1,axes2) = plt.subplots(nrows=1, ncols=2) # Axis is a numpy array, so we can assign it directly to the axes1 and axes2 arrays
axes1.plot(x, y, '--')
axes2.plot(y, x, 'r--')
```

```python
fig, axes = plt.subplots(2,4) 
#axes1.plot(x, y, '--')
#axes2.plot(y, x, 'r--')
axes[0, 0].plot(x, np.cos(x))
axes[0, 1].plot(x, np.sin(x), 'r')
axes[0, 2].plot(x, np.tan(x), 'y')
axes[0, 3].plot(x, np.cos(x)**2 + np.sin(x), 'g')
```

```python
fig, ((ax1,ax2,ax3,ax4), (ax5,ax6,ax7,ax8)) = plt.subplots(2,4) 
#axes1.plot(x, y, '--')
#axes2.plot(y, x, 'r--')
ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x), 'r')
ax3.plot(x, np.tan(x), 'y')
ax4.plot(x, np.cos(x)**2 + np.sin(x), 'g:')
ax5.plot(x, np.cos(x)*np.sin(x))
ax6.plot(x, np.cos(x)/np.tan(x), 'r--')
ax7.plot(x, np.cos(x)+np.sin(x)*np.tan(x), 'b--')
ax8.plot(x, np.cos(x)-np.sin(x)*np.tan(x), 'g-')
fig.tight_layout() #To improve the visual appearance of the figure
```
