```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'ro')
plt.show()
```

```python
plt.plot(x, y, 'bx')
plt.show()
```

```python
plt.plot(x,y, 'gD-')
plt.show()
```

```python
plt.plot(x,y, 'gs:')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x, y)
```

```python
plt.boxplot(x)
plt.show()
```
