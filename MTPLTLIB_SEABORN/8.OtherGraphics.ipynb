```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

# HISTOGRAM

```python
plt.hist(data, bins=10, histtype='step') #Histogram is to see the distribution of the data
#bins is the number of bars in the histogram
#we cannot use width and histtype
```

# BOXPLOT

```python
data2= np.append(data, 90)
plt.boxplot(data2, vert=False, patch_artist=True, notch=True, showfliers=True) #We can use false
#vert is to turn the boxplot, patch_artist is to fill the box, notch is to get thicker the center line
plt.show()
```
