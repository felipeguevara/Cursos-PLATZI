```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4)) #FIGURE IS THE ALL CANVAS, AXES ARE THE PARTS OF THE CANVAS
axes[0].plot(x, y, label='$sin(x)$') #The symbol $ help us to see the label in matematical notation
axes[0].set_title('Relacion X and Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend(loc='best')

axes[1].plot(y, x)
axes[1].set_title('Relacion Y and X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')

plt.tight_layout()
```

```python
plt.plot(x, y, label='sin(x)')
plt.title('This is the sin(x) function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right', frameon=False, bbox_to_anchor=(1.25, 0.5)) 
```
