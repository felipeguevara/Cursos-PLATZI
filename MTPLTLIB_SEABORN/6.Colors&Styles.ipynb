```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x**2, 'y.-')
ax.plot(x, x**1, 'b:')
ax.plot(x, x+3*x, 'w-')
ax.plot(x, x**3, 'r--')
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x**2, color='green', alpha=0.5, linewidth = 2, linestyle = '-', marker='o') 
# Alpha value is for transparency / linewidth is the thickness of the line
ax.plot(x, x**1, color='blue', alpha=0.8, linewidth = 3, linestyle = '-.', marker='d') 
# linestyle & marker are for customization of the line and the dots
ax.plot(x, x+3*3, color='#66FF66', alpha=0.3, linewidth = 4, linestyle = 'dashed', marker='x', markersize = 10, markerfacecolor = '#70FF70')
ax.plot(x, x**3, color='#78EEFF', alpha=0.6, linewidth = 5, linestyle = ':', marker='P')
```
