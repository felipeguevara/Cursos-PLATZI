```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['Colombia','Mexico','Germany','United Kingdom','United States']
population = [229312321,324234324,542445523,534353646,563423422]
```

```python
plt.bar(country,population, width=0.5, color = ['yellow','green','black']) #width is for the thickness of the bar
plt.xticks(np.arange(len(population)), ('COL','MEX','GER','UKG','USA'), rotation=45) 
plt.title('country population')
plt.xlabel('country')
plt.ylabel('population')
#first value is for the numbers of names of each bar, second for the aliases of each bar, third for the rotation
plt.show()
```

```python
plt.barh(country, population, color='blue', alpha=0.3)
plt.show()
```
