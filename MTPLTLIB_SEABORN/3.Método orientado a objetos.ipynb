```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x, y, 'ro')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9]) #The first two reference the position on the figure, the seonds reference the size of the axes
axes2 = fig.add_axes([0.1,0.6,0.3,0.3]) 
axes.plot(x, y, 'ro')
axes2.plot(y, x, 'b')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9]) #The first two reference the position on the figure, the seonds reference the size of the axes
axes2 = fig.add_axes([0.1,0.6,0.3,0.3]) 
axes.plot(x, y, 'ro')
axes2.plot(y, x, 'b')
axes2.set_facecolor('yellow')
fig.show()
```
